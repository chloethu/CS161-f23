# Dictionaries

## dict = {'key' : 'value', 'key2':'value2', etc}

```python
Butler = {'genre':'science fiction', 'award':'hugo'}
```

```python
# access info in the dictionary
print(Butler['genre'])
```

```python
print(Butler['award'])
```

```python
# adding info to a dict
Butler['birth_year'] = 1947
```

```python
print(Butler)
```

```python
Robinson = {}
```

```python
print(type(Robinson))
```

```python
Robinson['genre'] = 'science fiction'
```

```python
print(Robinson)
```

```python
Robinson['award'] = 'hugo'
```

```python
Robinson['birth year'] = 1952
```

```python
print(Robinson)
```

## Moving Through a Dictionary

```python
for key in Butler.keys():
    print(key)
```

```python
for v in Butler.values():
    print(v)
```

```python
for k, v in Butler.items():
    print(k, v)
```

## list of dictionaries

```python
sf_authors = [Butler, Robinson]
```

```python
print(sf_authors)
```

```python
print(sf_authors[0]['birth_year'])
```

## Word Counter

```python
import requests
```

```python
url = "https://raw.githubusercontent.com/msaxton/18th-century-historians/main/gibbon/gibbon_decline_and_fall/gibbon_decline_volume1_chap01.txt"
```

```python
response = requests.get(url)
```

```python
text = response.text
```

```python
text[:200]
```

```python
word_count = {}
```

```python
words = text.split()
```

```python
words[:100]
```

```python
for word in words:
    lower_word = word.lower()
    if lower_word in word_count.keys():
        word_count[lower_word] += 1
    else:
        word_count[lower_word] = 1
```

```python
sorted_word_counts = sorted(word_count.items(), key=lambda x:x[1], reverse=True)
```

```python
sorted_word_counts
```
